# Step 3: log Model

```python
import pandas as pd
import numpy as np
import warnings
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction import stop_words
from sklearn.metrics import confusion_matrix, roc_auc_score
import time
import matplotlib.pyplot as plt

pd.options.display.max_columns = 100
pd.options.display.max_rows = 3000
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
data['subreddit'].value_counts()
```

# Train, test data reform

```python
X = data['combined']
y = data['subreddit_n']
```

```python
X.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 42)
```

# logistic CountVectorizer

```python
pipe_cvec = Pipeline([
    ('cvec', CountVectorizer()),
    ('lr', LogisticRegression()),
])

params_cvec = {
    'cvec__stop_words' : [None,'english'],
    'cvec__ngram_range' :[(1,1),(1,2),(1,3)],
}

gs_cvec = GridSearchCV(pipe_cvec, 
                  params_cvec,
                  cv = 5) 
```

```python
t0 =time.time()
```

```python
gs_cvec.fit(X_train, y_train)
gs_cvec_model = gs_cvec.best_estimator_

# Score model on training set.
print (f'training accuracy: {gs_cvec_model.score(X_train, y_train)}')

# Score model on testing set.
print (f'testing accuracy: {gs_cvec_model.score(X_test, y_test)}')
```

```python
print (time.time() - t0)
```

```python
# best parameter
print (f'best parameter : {gs_cvec.best_params_}')
```

```python
preds = gs_cvec_model.predict(X_test)
```

```python
# distribution of outcome

pred_proba = [i[1] for i in gs_cvec_model.predict_proba(X_test)]

pred_proba_df = pd.DataFrame({
    'true' : y_test,
    'pred_prob' : pred_proba
})

# Adapt from notes Classification Metrics II

# Create figure.
plt.figure(figsize = (10,7))

# Create two histograms of observations.
plt.hist(pred_proba_df[pred_proba_df['true'] == 0]['pred_prob'],
         bins = 25,
         color = 'b',
         alpha = 0.6,
         label = 'Outcome = 0')
plt.hist(pred_proba_df[pred_proba_df['true'] == 1]['pred_prob'],
         bins = 25,
         color = 'orange',
         alpha = 0.6,
         label = 'Outcome = 1')

# Add vertical line at P(Outcome = 1) = 0.5.
plt.vlines(x = 0.5,
           ymin = 0,
           ymax = 1200,
           color = 'r',
           linestyle = '--')

# Add annotations for TN, FN, TP, FP.
plt.annotate(xy = (0.04, 1100),
             s = 'TN',
             size = 20)

plt.annotate(xy = (0.04, 200),
             s = 'FN',
             size = 20)

plt.annotate(xy = (1, 800),
             s = 'TP',
             size = 20)

plt.annotate(xy = (1, 50),
             s = 'FP',
             size = 20)

# Label axes.
plt.title('CountV Log Distribution of P(Outcome = 1, Pro-Trump)', fontsize = 22)
plt.ylabel('Frequency', fontsize = 18)
plt.xlabel('Predicted Probability that Outcome = 1 (Pro-Trump)', fontsize = 18)

# Create legend.
plt.legend(fontsize = 20);

# rows are actual, columns are prediction
tn,fn,fp,tp = confusion_matrix(y_test, preds).ravel()
cm = confusion_matrix(y_test, preds)
```

```python
# get confusion matrix
preds = gs_cvec_model.predict(X_test)

confusion_table_log = pd.DataFrame(cm, 
                               columns = ['predicted anti-trump posts','predicted pro-trump posts'], 
                               index = ['actual anti-trump posts','actual pro-trump posts'])
confusion_table_log
```

## logistic TFIDFVectorizer

```python
pipe_tfidf = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('lr', LogisticRegression(solver = 'saga')),
])

params_tfidf = {
    'tfidf__stop_words' : [None,'english'],
    'tfidf__ngram_range' :[(1,1),(1,2),(1,3)],
    'lr__penalty' : ['l1','l2'],
    'lr__C' : [0.1, 1, 10]
}

gs_tfidf = GridSearchCV(pipe_tfidf, 
                  params_tfidf,
                  cv = 5) 
```

```python
t1 = time.time()
```

```python
gs_tfidf.fit(X_train, y_train)
gs_tfidf_model = gs_tfidf.best_estimator_

# Score model on training set.
print (f'training accuracy: {gs_tfidf_model.score(X_train, y_train)}')

# Score model on testing set.
print (f'testing accuracy: {gs_tfidf_model.score(X_test, y_test)}')
```

```python
print (time.time()-t1)
```

```python
# best parameter
gs_tfidf.best_params_
```

```python
preds_tfidf = gs_tfidf_model.predict(X_test)
```
