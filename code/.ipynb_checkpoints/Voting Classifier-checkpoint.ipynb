```python
import pandas as pd
import numpy as np
import warnings
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, \
GradientBoostingClassifier, VotingClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction import stop_words
from sklearn.metrics import confusion_matrix, roc_auc_score
import time
import matplotlib.pyplot as plt

pd.options.display.max_columns = 100
pd.options.display.max_rows = 3000
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../datasets/data.csv')
```

```python
X = data['combined']
y = data['subreddit_n']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 42)
```

## Voting Classifer (Adaboost, GradientBoost, RandomForest) TFIDFVectorizer

```python
vc = VotingClassifier([
    ('ada', AdaBoostClassifier()),
    ('grad_boost', GradientBoostingClassifier()),
    ('random_forest', RandomForestClassifier()),
], voting='soft'
)



pipe = Pipeline([
    ('tfidf',TfidfVectorizer()),
    ('vc', vc)]
)
               
params = {
    'tfidf__ngram_range' : [(1,2)],
    'vc__grad_boost__n_estimators' : [100, 500, 1000],
    'vc__grad_boost__subsample' : [0.1,1],
}

gs_tfidf =GridSearchCV(pipe, params, cv=5)
```

```python
t0 = time.time()
gs_tfidf.fit(X_train, y_train)

# Score model on training set.
print (f'training accuracy: {gs_tfidf.score(X_train, y_train)}')

# Score model on testing set.
print (f'testing accuracy: {gs_tfidf.score(X_test, y_test)}')

print (time.time()-t0)
```

```python
gs_tfidf.best_score_
```

```python
gs_tfidf.best_params_
```

```python
pred_proba = [i[1] for i in gs_tfidf.predict_proba(X_test)]

pred_proba_df = pd.DataFrame({
    'true' : y_test,
    'pred_prob' : pred_proba
})

# Create figure.
plt.figure(figsize = (10,7))

# Create two histograms of observations.
plt.hist(pred_proba_df[pred_proba_df['true'] == 0]['pred_prob'],
         bins = 25,
         color = 'b',
         alpha = 0.6,
         label = 'Outcome = 0')
plt.hist(pred_proba_df[pred_proba_df['true'] == 1]['pred_prob'],
         bins = 25,
         color = 'orange',
         alpha = 0.6,
         label = 'Outcome = 1')

# Add vertical line at P(Outcome = 1) = 0.5.
plt.vlines(x = 0.5,
           ymin = 0,
           ymax = 1200,
           color = 'r',
           linestyle = '--')

# Add annotations for TN, FN, TP, FP.
plt.annotate(xy = (0.3, 1000),
             s = 'TN',
             size = 20)

plt.annotate(xy = (0.3, 160),
             s = 'FN',
             size = 20)

plt.annotate(xy = (0.8, 300),
             s = 'TP',
             size = 20)

plt.annotate(xy = (0.8, 50),
             s = 'FP',
             size = 20)

# Label axes.
plt.title('TFIDF Voting Distribution of P(Outcome = 1, Pro-Trump)', fontsize = 22)
plt.ylabel('Frequency', fontsize = 18)
plt.xlabel('Predicted Probability that Outcome = 1 (Pro-Trump)', fontsize = 18)

# Create legend.
plt.legend(fontsize = 20);

# rows are actual, columns are prediction
preds_tfidf = gs_tfidf.predict(X_test)

tn,fn,fp,tp = confusion_matrix(y_test, preds_tfidf).ravel()
cm = confusion_matrix(y_test, preds_tfidf)
```

```python
confusion_table_vote_tfidf = pd.DataFrame(cm, 
                               columns = ['predicted anti-trump posts','predicted pro-trump posts'], 
                               index = ['actual anti-trump posts','actual pro-trump posts'])
confusion_table_vote_tfidf
```
